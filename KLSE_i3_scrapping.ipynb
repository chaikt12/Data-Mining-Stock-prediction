```python
table=soup.find('tbody')
rows=table.find_all('tr')

for row in rows:
    cols=row.find_all('td')
    cols=[x.text.strip() for x in cols]
    pd.DataFrame(cols)
    print(cols)
```

```python
data=[]
table=soup.find('tbody')
rows=table.find_all('tr')

for row in rows:
    cols=row.find_all('td')
    cols=[x.text.strip() for x in cols]
    ann_date=cols[1]
    quarter=cols[2]
    revenue=cols[4]
    pbt=cols[5]
    np=cols[6]
    np_to_sh=cols[7]
    np_margin=cols[11]
    roe=cols[12]
    eps=cols[16]
    dps=cols[18]
    naps=cols[20]
    qoq=cols[22]
    yoy=cols[23]
    data.append([ann_date,quarter,revenue,pbt,np,np_to_sh,np_margin,roe,eps,dps,naps,qoq,yoy])
  
```

```python
pd.DataFrame(data,columns=['Annual_Date','Quarter_Date','Revenue','PBT','NP','NP_to_SH','NP_Margin','ROE','EPS','DPS','NAPS','Q0Q_Performance','YoY_performance'])
```
