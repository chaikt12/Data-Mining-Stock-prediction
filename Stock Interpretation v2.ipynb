# Top Glove

```python
topglov['Date']=pd.to_datetime(topglov['Date'],format='%b %d, %Y')
topglov=topglov.sort_values(by=['Date'])

topglov_df=topglov[['Date','Price']].set_index('Date')
topglov_return=topglov_df.pct_change()
#apply(lambda x: x/x[0])
```

```python
topglov['lower_tail']=(topglov['Open']-topglov['Low'])/(topglov['Open']-topglov['Price'])
topglov['upper_tail']=(topglov['Open']-topglov['High'])/(topglov['Open']-topglov['Price'])
```

```python
topglov
```

```python
#worst drop
topglov_return.idxmin()
```
